# Data Preprocessing:
## 1- Data Cleaning
####  Handle categorical variables using one-hot encoding

Most ML algorithms work with numbers. So we will convert the attributes.

```python
df = pd.get_dummies(data, columns=['Geography', 'Gender', 'Card Type'])
```

```python
df.corr()
```

## 2-  Feature engineering 
### -feature selection-


#### Remove irrelevant columns 
based on corrlation

```python
df = df.drop(['Satisfaction Score','Tenure','NumOfProducts','HasCrCard',
              'Point Earned','Geography_France','Geography_Germany','Geography_Spain','Gender_Female','Gender_Male','Card Type_DIAMOND',
              'Card Type_GOLD','Card Type_PLATINUM','Card Type_SILVER','EstimatedSalary'
              ,'RowNumber', 'CustomerId', 'Surname'], axis=1)
df
```

## 3- Normalizition - Min Max Scaler

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
X
```

```python
mm_scaler = preprocessing.MinMaxScaler()
X_mm = mm_scaler.fit_transform(X)
```

# Spliting the data 
## 20% test 80% train

```python
X_train, X_test, y_train, y_test = train_test_split(X_mm, y, test_size=0.2, random_state=42)
```

# Applying machine learning algorthim
## First model
### K Neighbors Classifier  -baseline-
initialize the model

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

Fit the model on the training data

```python
knn.fit(X_train,y_train)
```

 Make predictions on the test data

```python
y_pred = knn.predict(X_test)
```

### Evaluate

Calculate the accuracy of the model:

```python
print('Training Accuracy : {:.3f}'.format(knn.score(X_train, y_train)))
print('Test Accuracy : {:.3f}'.format(knn.score(X_test, y_test))) 
```

Precision

```python
precision = precision_score(y_test, y_pred)
print('Precision:', precision)
```

Recall

```python
recall = recall_score(y_test, y_pred)
print('Recall:', recall)
```

ROC curve 

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
# Compute fpr, tpr, thresholds and roc auc
y_score = knn.predict_proba(X_test)[::,1]
fpr, tpr, thresholds = roc_curve(y_test,y_score)
roc_auc = roc_auc_score(y_test,y_score)

# Plot ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('KNeighbors Model')
plt.legend(loc="lower right")
```
